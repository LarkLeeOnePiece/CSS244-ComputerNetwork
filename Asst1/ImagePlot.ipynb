```python
import re

def extract_bandwidth(text):
    # extract Bandwidth (get Mbits/sec data)
    #pattern = r'(\d+\.+\d+)\s+Mbits/sec' #for float
    pattern = r'(\d+)\s+Mbits/sec' #for int
    bandwidths = re.findall(pattern, text)
    
    # return an array
    return [float(bw) for bw in bandwidths]

# demo
text_data = """
[  4]   0.00-1.00   sec  78.6 MBytes   659 Mbits/sec
[  4]   1.00-2.00   sec  78.4 MBytes   657 Mbits/sec
[  4]   2.00-3.00   sec  80.0 MBytes   672 Mbits/sec
[  4]   3.00-4.00   sec  80.4 MBytes   674 Mbits/sec
[  4]   4.00-5.00   sec  80.2 MBytes   674 Mbits/sec
[  4]   5.00-6.00   sec  80.4 MBytes   674 Mbits/sec
[  4]   6.00-7.00   sec  80.4 MBytes   675 Mbits/sec
[  4]   7.00-8.00   sec  80.2 MBytes   674 Mbits/sec
[  4]   8.00-9.00   sec  80.4 MBytes   674 Mbits/sec
[  4]   9.00-10.00  sec  80.4 MBytes   676 Mbits/sec
[  4]  10.00-11.00  sec  80.0 MBytes   671 Mbits/sec
[  4]  11.00-12.00  sec  80.2 MBytes   673 Mbits/sec
[  4]  12.00-13.00  sec  80.0 MBytes   672 Mbits/sec
[  4]  13.00-14.00  sec  80.4 MBytes   674 Mbits/sec
[  4]  14.00-15.00  sec  80.6 MBytes   677 Mbits/sec
[  4]  15.00-16.00  sec  80.4 MBytes   675 Mbits/sec
[  4]  16.00-17.00  sec  80.8 MBytes   678 Mbits/sec
[  4]  17.00-18.00  sec  80.0 MBytes   671 Mbits/sec
[  4]  18.00-19.00  sec  79.8 MBytes   671 Mbits/sec
[  4]  19.00-20.00  sec  80.0 MBytes   671 Mbits/sec
- - - - - - - - - - - - - - - - - - - - - - - - -
[ ID] Interval           Transfer     Bandwidth
[  4]   0.00-20.00  sec  1.56 GBytes   672 Mbits/sec                  sender
[  4]   0.00-20.00  sec  1.56 GBytes   672 Mbits/sec                  receiver             receiver                receiver                 receiver               receiver                receiver
"""

# extract the bandwith
bandwidth_data = extract_bandwidth(text_data)
if len(bandwidth_data)!=22:
    print("len=",len(bandwidth_data))
    print(bandwidth_data)
else:
    print(f"bandwidth_data={bandwidth_data}")
    print(f"\n|  | 1s | 3s | 5s | 7s | 9s | 12s | 15s | 20s | AVG |\n| --- | --- | --- | --- | --- | --- | --- | --- | --- | --- |\n| Bandwidth | {bandwidth_data[0]} Mbits/s | {bandwidth_data[2]} Mbits/s | {bandwidth_data[4]} Mbits/s | {bandwidth_data[6]} Mbits/s | {bandwidth_data[8]} Mbits/s | {bandwidth_data[10]} Mbits/s | {bandwidth_data[14]} Mbits/s | {bandwidth_data[19]} Mbits/s |{bandwidth_data[20]} Mbits/s |")
import matplotlib.pyplot as plt

def plot_bandwidth_array(bandwidth_array):
    # get the x
    x = range(len(bandwidth_array)-2)
    
    # plot
    plt.figure(figsize=(10, 6))
    plt.plot(x, bandwidth_array[0:-2], marker='o', linestyle='-', color='b')
    plt.xlabel('Time(s)')
    plt.ylabel('Bandwidth (Mbits/sec)')
    plt.title('Bandwidth vs Time')
    # 
    plt.xticks(x)
    plt.grid(True)
    plt.show()

# plot
bandwidth_data = bandwidth_data
plot_bandwidth_array(bandwidth_data)
```
