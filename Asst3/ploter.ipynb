## 1.1 TCP wired test

```python
## Here we get the data and average it
import os 
import json
import numpy as np
import time
ROUND=4
time_s=50
txqlen=[1,4,8,12,16,20,25,26,30,34,40,50,100,200,800,3000,4000,5000,6000,7000,8000,9000,10000]
#txqlen=[1,4,8,12,16,20,25,26,30,34,40,50,60,70,80,90,100,200,800,12000,2000,3000,4000]# for udp test
expo=f"test_{time_s}"
sleeptime=1
proto="tcp"
net=""#or"wireless""
_out=f"{expo}_{proto}"
# 定义iperf3令
txqlen_bw_dict={}
txqlen_rtt_dict={}
for txql in txqlen:
        txqlen_bw_dict[str(txql)]=[]
        txqlen_rtt_dict[str(txql)]=[]
        for r in range(ROUND):
                test_out=f"data/{_out}/output_txql_{txql}_r_{r}.json"
                if os.path.exists(test_out):
                    with open(test_out, 'r') as reno_file:
                        data = json.load(reno_file) 
                    intervals = data['intervals']
                    for interval in intervals:
                        throughput = interval['streams'][0]['bits_per_second']
                        rtt = interval['streams'][0]['rtt']
                        txqlen_bw_dict[str(txql)].append(throughput/1000000)
                        txqlen_rtt_dict[str(txql)].append(rtt/1000)#ms
                else:
                    print(f"The file {test_out} does not exist.")
        print(f"txql({txql}):{len(txqlen_bw_dict[str(txql)])}")
                
```

### 1.1.1 TCP througput plot

```python
import numpy as np
import matplotlib.pyplot as plt

# 示例字典，每个键对应一个列表
A = txqlen_bw_dict

# 计算每个键对应列表的平均值
averages = {k: np.mean(v) for k, v in A.items()}

# 提取键和对应的平均值用于绘图
keys = list(averages.keys())
values = list(averages.values())

# 创建条形图
plt.figure(figsize=(12, 6))
bars = plt.bar(keys, values, color='skyblue')

# 在条形图的每个条形上添加数值标签
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), va='bottom')  # va: vertical alignment


plt.tight_layout()  # 自动调整子图参数，使之填充整个图像区域


# 设置图表标题和轴标签等
plt.title(f'Wired network average throuput over queue length({proto})')
plt.xlabel('Queue length')
plt.ylabel('Througput(Mbytes/s)')

# 显示图形
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 假设 data 是一个形状为 [size, N] 的 numpy 数组
# 例如：
data =np.array(throughput_list)

# 计算整个数组的平均值
overall_mean = np.mean(data)

# 创建条形图
fig, ax = plt.subplots()
bar = ax.bar('Average', overall_mean, color='blue')

# 在条形图的顶部添加数值标签
ax.annotate(f'{overall_mean:.2f}',
            xy=(0, overall_mean),
            xytext=(0, 3),  # 3 points vertical offset
            textcoords="offset points",
            ha='center', va='bottom')

# 设置图表标题和轴标签等
ax.set_title('Overall Average Value of Array')
ax.set_ylabel('Average Value')

# 显示图形
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 创建一个随机的三维数组 [P, size, N]
# 例如，P = 10, size = 5, N = 3
P, size, N = 10, 5, 3
data = np.random.rand(P, size, N)

# 计算每个 [size, N] 数组的平均值
means = np.mean(data, axis=(1, 2))  # 沿着 size 和 N 维度进行平均

# 创建条形图
fig, ax = plt.subplots()
bar_positions = np.arange(P)  # 为每个平均值生成一个条形位置
bars = ax.bar(bar_positions, means, color='skyblue')

# 在条形图的每个条形上添加数值标签
for bar in bars:
    height = bar.get_height()
    ax.annotate(f'{height:.2f}',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3),  # 3 points vertical offset
                textcoords="offset points",
                ha='center', va='bottom')

# 设置图表标题和轴标签等
ax.set_title('Average Values of Each [size, N] Sub-array')
ax.set_xlabel('Sub-array Index')
ax.set_ylabel('Average Value')
plt.xticks(bar_positions)  # 设置x轴刻度标签为条形的位置

# 显示图形
plt.show()
```
